# HFT Platform Research Environment

Welcome to the research environment. Here you can:
1. Query historical market data from ClickHouse.
2. Analyze order execution.
3. Run backtests using `hftbacktest`.

### Setup Connection

```python
import clickhouse_connect

client = clickhouse_connect.get_client(host='clickhouse', port=8123, username='default', password='')
print('Connected to ClickHouse')
```

### Query Market Data
Fetch the LOB snapshots and Ticks for a symbol.

```python
query = """
SELECT toDateTime(exch_ts / 1e9) as ts, price, volume, type
FROM hft.market_data
WHERE symbol = '2330'
ORDER BY exch_ts DESC
LIMIT 100
"""
df = client.query_df(query)
df.head()
```
